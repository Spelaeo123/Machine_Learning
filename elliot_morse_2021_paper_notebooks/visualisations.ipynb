# final visualisations for paper

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

# tsne bedrock

```python
save_plots = True
```

```python
filepath = 'final_figures/bedrock_tsne.eps'
tsne_data_with_colour = pd.read_csv('output_datasets/pre_model_datasets/tsne_data_with_color_bedrock.csv')
```

```python
my_pallete = ['#e6194b', '#3cb44b', '#ffe119', '#4363d8', '#f58231', '#911eb4', '#46f0f0', '#f032e6', '#bcf60c', '#fabebe', '#008080', '#9a6324', '#fffac8', '#800000', '#aaffc3', '#808000', '#ffd8b1', '#000075', '#808080', '#000000']
```

```python
sns.set(rc={'figure.figsize':(4,4)})
sns.set_style("whitegrid",{'grid.linestyle': 'dashed'})
sns.set_palette(sns.color_palette(my_pallete, 19))

plot_on_x = 'tsne1'
plot_on_y = 'tsne2'

plot = sns.scatterplot(data = tsne_data_with_colour, linewidth = 0, edgecolor = 'black', s= 10, x = plot_on_x, y = plot_on_y, hue='class') #palette=sns.color_palette("colorblind", tsne_data_with_colour['class'].nunique())

plt.title('t-SNE Analysis of Bedrock Sample Data', fontsize = 12)

plt.yticks(fontsize = 10)
plt.xticks(fontsize = 10)

plot.set_xlabel('tSNE1',fontsize=10)
plot.set_ylabel('tSNE2',fontsize=10)

plt.legend(fontsize=10, markerscale=0.5, bbox_to_anchor=(1.02, 1), loc=2, borderaxespad=0.)

if save_plots == True:
    fig = plot.get_figure()
    fig.savefig(fname=filepath, dpi = 2000, pad_inches = 0.5, bbox_extra_artists=(plot,), bbox_inches='tight')
```

# tsne outliers

```python
filepath = 'final_figures/outlier_tsne.eps'

tsne_data_with_colour = pd.read_csv('output_datasets/pre_model_datasets/tsne_outlier_data.csv')
```

```python
tsne_data_with_colour.columns
```

```python
def outlier_label(row):
    if row['inlierLabel'] == -1:
        return('outlier')
    elif row['inlierLabel'] == 1:
        return('inlier')
```

```python
tsne_data_with_colour['class'] = tsne_data_with_colour.apply(outlier_label, axis = 1)
```

```python
tsne_data_with_colour.head()
```

```python
sns.set(rc={'figure.figsize':(4,4)})
sns.set_style("whitegrid", {'grid.linestyle': 'dashed'})

plot_on_x = 'tsne1'
plot_on_y = 'tsne2'

plot = sns.scatterplot(data = tsne_data_with_colour, s= 10, x = plot_on_x, y = plot_on_y, hue='class', linewidth = 0, edgecolor = 'black',
                palette=sns.color_palette("muted", tsne_data_with_colour['class'].nunique()))

plt.title('t-SNE of Artefact Sample Data', fontsize = 12)

plt.legend(fontsize=10, markerscale=0.5)

plt.yticks(fontsize = 10)
plt.xticks(fontsize = 10)

plot.set_xlabel('tSNE1',fontsize=10)
plot.set_ylabel('tSNE2',fontsize=10)

plt.tight_layout()

if save_plots == True:
    fig = plot.get_figure()
    fig.savefig(fname=filepath, dpi = 2000, pad_inches = 0.5)
```

# learning curve

```python
filepath = 'final_figures/LC.eps'

rfc_lc = pd.read_csv('output_datasets/pre_model_datasets/rfc_lc.csv')
```

```python
sns.set(rc={'figure.figsize':(4,4)})
sns.set(rc={"lines.linewidth": 1})
sns.set_style("whitegrid", {'grid.linestyle': 'dashed'})

plot = sns.lineplot(x = 'train_sizes', y = 'k_fold_stratified_scores', hue = 'train_or_test_data', ci = 'sd', err_style='bars', data = rfc_lc, markers = True)

plot.set_ylabel("F1 score",fontsize=10)
plot.set_xlabel("Train fold size (number of observations)",fontsize=10)
plot.tick_params(labelsize=10, length=8, width=2 )

plt.title('Learning Curve', fontsize = 10)

plt.yticks(fontsize = 10)
plt.xticks(fontsize = 10)

plt.tight_layout()

plt.legend(fontsize = 10, loc = 4)

if save_plots == True:
    fig = plot.get_figure()
    fig.savefig(filepath)
```

# knn class F1 scores

# knn f1 vs sample size

```python
forPlot = pd.read_csv('output_datasets/knn_datasets/score_vs_sample_number.csv')
filepath = 'final_figures/f1vssamplesizeknn.eps'
```

```python
sns.set(rc={'figure.figsize':(4,4)})
sns.set(font_scale=0.6)
sns.set_style("whitegrid", {'grid.linestyle': 'dashed'})

plot = sns.scatterplot(x ='Number of Observations', y = 'Mean F1 Score', markers='class', data = forPlot, s = 14)# alpha = 0.2)

plt.title("KNN", fontsize = 10)

plt.yticks(fontsize = 10)
plt.xticks(fontsize = 10)
plt.tight_layout()
plot.set_ylabel('F1 score', fontdict={'fontsize': 10})
plot.set_xlabel("Number of Observations", fontdict={'fontsize': 10})

for line in range(0, forPlot.shape[0]):    
    plot.text((int(forPlot['Number of Observations'].iloc[line]) + 1.5), (float(forPlot['Mean F1 Score'].iloc[line])+0.01), forPlot['class'].iloc[line], horizontalalignment='left', size=8, color='black', weight='regular')

if save_plots:
    fig = plot.get_figure()
    fig.savefig(filepath)
```
