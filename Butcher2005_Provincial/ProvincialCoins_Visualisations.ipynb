## Save Plot switch

```python
save_plots = False
```

## PCA Visualisation

```python
filepath = 'Figure/Butcher2005_ProvincialCoins_Tom_PCA_Emperor.png'
```

```python
my_pallete = ['#e6194b', '#3cb44b', '#ffe119', '#4363d8', '#f58231', '#911eb4', '#46f0f0', '#f032e6', '#bcf60c', '#fabebe', '#008080', '#9a6324', '#fffac8', '#800000', '#aaffc3', '#808000', '#ffd8b1', '#000075', '#808080', '#000000']
```

```python
sns.set(rc={'figure.figsize':(10,10)})
sns.set_palette(sns.color_palette(my_pallete, 19))

plot_on_x = 'PC1'
plot_on_y = 'PC2'

plot = sns.scatterplot(data = EmpCol_RefCol_PCAs, linewidth = 0.3, edgecolor = 'black', s= 5, x = plot_on_x, y = plot_on_y, hue='EMPEROR',
                #palette=sns.color_palette("colorblind", tsne_data_with_colour['class'].nunique())
                      )

plt.title('PCA Plot of Coins', fontsize = 10)

plt.yticks(fontsize = 5)
plt.xticks(fontsize = 5)

plot.set_xlabel('PC1',fontsize=7)
plot.set_ylabel('PC2',fontsize=7)


plt.legend(fontsize=7, markerscale=0.5, bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

if save_plots == True:
    fig = plot.get_figure()
    fig.savefig(fname=filepath, dpi = 2000, pad_inches = 0.5, bbox_extra_artists=(plot,), bbox_inches='tight')
    
```

## t-SNE Visualisation

```python
#filepath = 'Figure/Coin_tsne.eps'
filepath = 'Figure/Butcher2005_ProvincialCoins_Tom_tsne_Emperor.png'
```

```python
my_pallete = ['#e6194b', '#3cb44b', '#ffe119', '#4363d8', '#f58231', '#911eb4', '#46f0f0', '#f032e6', '#bcf60c', '#fabebe', '#008080', '#9a6324', '#fffac8', '#800000', '#aaffc3', '#808000', '#ffd8b1', '#000075', '#808080', '#000000']
```

```python
sns.set(rc={'figure.figsize':(15,15)})
sns.set_palette(sns.color_palette(my_pallete, 14))

plot_on_x = 'tsne1'
plot_on_y = 'tsne2'

plot = sns.scatterplot(data = tsne_df_EmpCol_comb, linewidth = 0.3, edgecolor = '', s= 20, x = plot_on_x, y = plot_on_y, hue='EMPEROR',
                #palette=sns.color_palette("colorblind", tsne_data_with_colour['class'].nunique())
                      )

plt.title('t-SNE Analysis of Coins', fontsize = 10)

plt.yticks(fontsize = 10)
plt.xticks(fontsize = 10)

plot.set_xlabel('tSNE1',fontsize=10)
plot.set_ylabel('tSNE2',fontsize=10)


plt.legend(fontsize=10, markerscale=0.5, bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)


fig = plot.get_figure()
fig.savefig(fname=filepath, dpi = 2000, pad_inches = 0.5, bbox_extra_artists=(plot,), bbox_inches='tight')
```
